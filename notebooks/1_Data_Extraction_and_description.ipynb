**Let us extract category and sub category**

```python
def get_cat(x):
    dic = eval(x)
    cat = dic['slug'].split("/")
    return cat[0]

def get_subcat(x):
    try:
        dic = eval(x)
        cat = dic['slug'].split("/")
        return cat[1]
    except:
        return "sub_"+str(cat[0])
```

```python
df['cat'] = df['category'].apply(get_cat)
df['sub_cat'] = df['category'].apply(get_subcat)
```

- As we can see there are many stated project can exist, but given the context of our problem successful and failed are the only two relevant states so let us filter 

```python
df['state'].value_counts()
```

```python
def getValidState(x):
    try:
        if(x=='successful' or x=='failed'):
            return True
        else:
            return False
    except:
        return False
```

```python
df = df[df['state'].apply(getValidState)]
```

- Launched_at, deadline and state_changed seemed to be dates in the unix time let us convert them into date formats

```python
from datetime import datetime
def get_date_format(x):
    ts = int(x)
    return (datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d'))
```

```python
df['launched_at'] = pd.to_datetime(df['launched_at'].apply(get_date_format))
df['deadline'] = pd.to_datetime(df['deadline'].apply(get_date_format))
df['state_changed_at'] = pd.to_datetime(df['state_changed_at'].apply(get_date_format))
df['created_at'] =  pd.to_datetime(df['created_at'].apply(get_date_format))
```
