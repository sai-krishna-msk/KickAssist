## Hypothesis Testing
*Validating if our hypothesis(Assumptions) from above observation about the data are statistically true*


**Let us test of campaigns which are picked by staff_picked failed and failed in general have significantly different pledged amounts, It would prove that being staff_picked is still valuable even though kickstarter has a policy of all or nothing, because if more people have pledged it means more people have liked our product**


```python
from scipy.stats import ttest_ind
failed = df[df['status']=='failed']['pledged']
stf0_failed = df[(df['status']=='failed') & (df['staff_pick']==False)]['pledged']
stat, p = ttest_ind(failed, stf0_failed)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distributions')
```

*Distribution is significantly different perhaps we could be really confident if p value was less than 1%*

**Let us check if days for deadline(Number of days before campaign) ends is significantly different for successful and failed projects**

```python
from scipy.stats import ttest_ind
failed = df[df['status']=='failed']['days_to_deadline']
success = df[df['status']=='successful']['days_to_deadline']
stat, p = ttest_ind(failed, success)
print('stat=%.3f, p=%.5f' % (stat, p))
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distributions')
```

*days to deadline is in deed a predictable feature and we can more sure of it than above assertion(because of p value)*


**Let us check if amount which we get more than our goal, is significantly different from successful campaigns which are staff picked not picked and overall**


```python
from scipy.stats import f_oneway
df['beyound_goal'] = df['pledged']-df['goal']
stf_success = df[  (df['status']=='successful') & (df['staff_pick']==True)]['beyound_goal']
Nstf_success = df[ (df['status']=='successful') & (df['staff_pick']==False)]['beyound_goal']
by_success = df[df['status']=='successful']['beyound_goal']
stat, p = f_oneway(stf_success , Nstf_success , by_success)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distributions')
print(f"Staff_pick:1 successful: {stf_success.mean()}")
print(f"Staff_pick:0 successful: {stf0_failed.mean()}")
print(f"Successful whole-data: {by_success.mean()}")
```

*With the help of above test we can conclude that irrespective of our project being successfully funded or not being staff picked gives your project more popularity*

```python
# Python program to generate WordCloud 

# importing all necessery modules 
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 

def create_word_cloud(words, title):
    comment_words = ' '
    stopwords = set(STOPWORDS) 
    stopwords.update(set(['failed']))
    
    for val in words: 
        val = str(val) 
        tokens = val.split() 
        for i in range(len(tokens)): 
            tokens[i] = tokens[i].lower() 
            
        for words in tokens: 
            comment_words = comment_words + words + ' '
            
    wordcloud = WordCloud(width = 800, height = 800, 
                              background_color ='white', 
                              stopwords = stopwords, 
                              min_font_size = 10).generate(comment_words) 
        
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.title(title)
    plt.axis("off") 
    plt.tight_layout(pad = 0) 
    plt.show() 
```

## Temporal Patterns
**Let us verify weather patterns and relationship's between our features and target variable have temporal component to them by which i mean if current years data is more closely related to previous three years rather than previous 6 years**
**This insight can be extremely valuable due to following reasons**
- It's not the case that more data equals greater accuracy, because if our data has temporal aspect then most of the data we would be using for model might be stale
- It takes more time to train model with more data
- If we would like to reuse the model for future years to come then every year we would be aggregating more and more useless data

### Brute-Force Modeling 


```python
from sklearn.ensemble import RandomForestClassifier
import operator


def score(X_train, X_test, y_train, y_test):
    rf_fet = {}
    from sklearn.ensemble import RandomForestClassifier
    rf= RandomForestClassifier(n_estimators=5, random_state=13579, n_jobs=-1)
    rf.fit(X_train, y_train)
    rf_score = rf.score(X_test, y_test)
    
   
    feat_labels = X_train.columns.values
    
    for feature, acc in zip(feat_labels, rf.feature_importances_):
        rf_fet[feature] = acc
        
    rf_fet =  sorted(rf_fet.items(), key=operator.itemgetter(1), reverse=True)
  
        
    return (rf,rf_score, rf_fet)



```

```python
cols = ['launched_at', 'deadline', 'status', 'days_to_deadline', 'goal',
       'sub_category', 'category', 'blurb_length', 'location_country', 'delivery', 'rewards_mean', 'rewards_median',
       'rewards_variance', 'rewards_SD', 'rewards_MIN', 'rewards_MAX' , 'rewards_NUM', 'currency', 'launch_year', 'launch_month', 'deadline_year', 'deadline_month']
df = df[cols]
df= df.dropna(axis=0, subset=["rewards_MIN"])
df= df.dropna(axis=0, subset=["blurb_length"])
df.sort_values("launched_at" , inplace=True)

df = df.reset_index(drop=True)


df["delivery"]  = pd.to_datetime(df["delivery"]).dt.date
df["launched_at"]  = pd.to_datetime(df["launched_at"]).dt.date
df.sort_values("launched_at" , inplace=True)

df.drop(['launched_at', 'delivery', 'deadline'] ,axis=1 , inplace=True)
df.reset_index(inplace=True)
df.drop('index', inplace=True , axis=1)



from sklearn.preprocessing import LabelBinarizer

binarizer= LabelBinarizer()
df["status"] = binarizer.fit_transform(df["status"])


from sklearn.preprocessing import OneHotEncoder
encoder  = OneHotEncoder(sparse=False)
cat_cols=['category', 'sub_category', 'currency', 'location_country']
X_hot = encoder.fit_transform(df[cat_cols])


onehotcols = []
for cat in encoder.categories_:
    for col in cat:
        onehotcols.append(col)

X_hot = pd.DataFrame(X_hot , columns=onehotcols)
df =pd.concat([df , X_hot] , axis=1)

df.drop(cat_cols , axis=1 , inplace=True)



```

### Modeling-2018

```python
train_years_ =[2009, 2010 , 2011, 2012 , 2013 , 2014 ,2015, 2016, 2017]
valid_years = [2018]
res_dict ={}
for i in range(len(train_years_)):
    
    train_years = train_years_[i:9]
    
    df_train = df[df['launch_year'].apply(lambda x: True if x in train_years else False)]
    df_valid= df[df['launch_year'].apply(lambda x: True if x in valid_years else False)]

    X_train , y_train = df_train.drop("status" , axis=1) , df_train['status']
    X_valid , y_valid = df_valid.drop("status" , axis=1) , df_valid['status']
    
    _ , scores , _ = score(X_train , X_valid , y_train , y_valid)
    trn_yrs = f"{train_years[0]}"+"-"+f"{train_years[-1]}"
    res_dict[trn_yrs] = scores
res_dict
      
```
