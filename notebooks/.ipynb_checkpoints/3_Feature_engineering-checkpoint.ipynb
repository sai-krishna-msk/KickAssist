- As we can see rewards column is present as array, but while training the model we can not give array as a value for one column, we have a choice to make here
    - We can add n number of columns to our main dataset where n is the maximum number of rewards the whole dataset has, so for projects with n-1 rewards, we can fill the nth column with zero
    - We can add columns representing the summary statistics of the data, like we can add a column representing the mean of the rewards , one for variance., for max etc
    
**Second option makes more sense to me, so let's try the second one out**
- As we have not consider currency for goal (instead we decided to add currency as a column) so to be consistent we shall ignore currency over here as well

```python
import re

def getMean(x):
  

  if(x=='[]'):
    return np.nan
  else:
    
    lis=[]
    res = x.strip('][').split(', ') 
    
    for each in res:
        lis.append(int(re.sub('[^0-9]+', '', each)))

    return np.array(lis).mean()

def getMedian(x):
    if(x=='[]'):
        return np.nan
    else:
        lis=[]
        res = x.strip('][').split(', ') 
        for each in res:
            lis.append(int(re.sub('[^0-9]+', '', each)))
            
    return np.median(lis)

def getVariance(x):
    if(x=='[]'):
        return np.nan
    
    else:
        lis=[]
        res = x.strip('][').split(', ') 
        for each in res:
            lis.append(int(re.sub('[^0-9]+', '', each)))
    return np.array(lis).var()


def getSD(x):
    if(x=='[]'):
        return np.nan
    
    else:
        lis=[]
        res = x.strip('][').split(', ') 
        for each in res:
            lis.append(int(re.sub('[^0-9]+', '', each)))
    return np.std(lis)

def getMIN(x):
    if(x=='[]'):
        return np.nan
    
    else:
        lis=[]
        res = x.strip('][').split(', ') 
        for each in res:
            lis.append(int(re.sub('[^0-9]+', '', each)))
    return min(lis)


def getMAX(x):
    if(x=='[]'):
        return np.nan
    
    else:
        lis=[]
        res = x.strip('][').split(', ') 
        for each in res:
            lis.append(int(re.sub('[^0-9]+', '', each)))
    return max(lis)




def getNUM(x):
    if(x=='[]'):
        return np.nan
    
    else:
        lis=[]
        res = x.strip('][').split(', ') 
   
    return len(res)
```

```python
df["rewards_mean"] = df["rewards"].apply(getMean)
df["rewards_median"] = df["rewards"].apply(getMedian)
df["rewards_variance"] = df["rewards"].apply(getVariance)
df["rewards_SD"] = df["rewards"].apply(getSD)
df["rewards_MIN"] = df["rewards"].apply(getMIN)
df["rewards_MAX"] = df["rewards"].apply(getMAX)
df["rewards_NUM"] = df["rewards"].apply(getNUM)
```

```python
df.drop(['rewards', "Unnamed: 0", "Unnamed: 0.1"], axis=1 , inplace=True)
```

- This seems messed up, null values are still fine but, there are some negative  values indicating rewards were estimated to be delivered before the project even started that does not make sense
- After further investigation (cross checking with couple of projects with negative values on the web and my scraping scripts i could not come to a concrete conclusion as to why it happened may be the creator of the campaign changed or something what ever might be the reason let us replace null values and negative values with different encodings 
- Telling the model not to treat both of them the same way

```python
def get_delivery_binary(x):
    
    if(pd.isna(x)):
        return -1
    elif(x<0):
        return 0
    else:
        return x



df_final['days_to_delivery'] = df_final['days_to_delivery'].apply(get_delivery_binary)
```
